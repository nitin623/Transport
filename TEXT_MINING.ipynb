```python
import nltk
nltk.download()
```

```python
df_true['description'] = df_true['description'].astype(str).str.lower()
df_true.head(5)
```

```python
from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

df_true['text_token']=df_true['description'].apply(regexp.tokenize)
df_true.head()
```

```python
from nltk.corpus import stopwords
# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")

# Extend the list with your own custom stopwords
my_stopwords = ['also','made','makes','like','this','even','company']
stopwords.extend(my_stopwords)
```

```python
df_true['text_token'] = df_true['text_token'].apply(lambda x: [item for item in x if item not in stopwords])
df_true.head()
```

```python
df_true['text_string'] =df_true['text_token'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
```

```python
df_true[['description', 'text_token', 'text_string']].head()
```

```python
all_words = ' '.join([word for word in df_true['text_string']])
```

```python
tokenized_words = nltk.tokenize.word_tokenize(all_words)
```

```python
from nltk.probability import FreqDist

fdist = FreqDist(tokenized_words)
fdist
```

```python
df_true['text_string_fdist'] = df_true['text_token'].apply(lambda x: ' '.join([item for item in x if fdist[item] >= 1 ]))
```

```python
df_true[['description', 'text_token', 'text_string', 'text_string_fdist']].head()
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lem = WordNetLemmatizer()

df_true['text_string_lem'] = df_true['text_string_fdist'].apply(wordnet_lem.lemmatize)
```

```python
df_true['is_equal']= (df_true['text_string_fdist']==df_true['text_string_lem'])
```

```python
df_true
```

```python
all_words_lem = ' '.join([word for word in df_true['text_string_lem']])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=600, 
                     height=400, 
                     random_state=2, 
                     max_font_size=100).generate(all_words_lem)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

words = nltk.word_tokenize(all_words_lem)
fd = FreqDist(words)
```

```python
# Obtain top 10 words
top_10 = fd.most_common(10)

# Create pandas series to make plotting easier
fdist = pd.Series(dict(top_10))
```

```python
fd.most_common(10)
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

sns.barplot(y=fdist.index, x=fdist.values, color='blue');
```
